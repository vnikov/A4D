```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
AGraph = nx.Graph()
```

```python
Nodes = range(1,5)
Edges = [(1,2), (2,3), (3,4), (4,5), (1,3), (1,5)]
```

```python
AGraph.add_nodes_from(Nodes)
AGraph.add_edges_from(Edges)
```

```python
nx.draw(AGraph, with_labels=True)
```
